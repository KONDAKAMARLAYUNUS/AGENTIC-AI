```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('wordnet', quiet=True)

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

words = ['running', 'studies', 'better', 'caring', 'happily']

for word in words:
    stem = stemmer.stem(word)
    lemma = lemmatizer.lemmatize(word, pos='v')

    print(f'{word:12} | stem: {stem:10} | lemma: {lemma}')
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download(['stopwords','wordnet','punkt'],quiet=True)
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()

def preprocess(text):
  text=text.lower()
  text=re.sub(r'[^a-zA-Z]',' ',text)
  tokens =text.split()
  filtered_tokens=[]
  for w in tokens:
    if w not in stop_words:
      filtered_tokens.append(w)

    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

  return ' '.join(lemmatized_tokens)
sample_text = 'i am going to running the some one story'
processed_text = preprocess(sample_text)
print(f"Original: {sample_text}\nProcessed: {processed_text}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
emails=[
    'cricket match cricket match cricket',
    'free win money free win money',
    'cricket match free win money cricket',
]
cv=CountVectorizer()
x=cv.fit_transform(emails).toarray()
print('vocabulary:',cv.get_feature_names_out())
print('row matrix:')
print(x)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
emails=[
    'cricket match cricket match cricket',
    'free win money free win money',
    'cricket match free win money cricket',
]
tfidf=TfidfVectorizer(sublinear_tf=True)
x=tfidf.fit_transform(emails).toarray()
print('feature:',tfidf.get_feature_names_out())
print('if-idf metrix(rounded):')

import numpy as np
print(np.round(x,3))
```
